```python
import json
import folium

m = folium.Map(location=[42.36055,-71.0606], zoom_start=15, control_scale=True)
folium.Marker(
    location=[42.3570104,-71.0710964],
    popup='Source',
    icon=folium.Icon(color='green', icon='ok-sign'),
).add_to(m)
folium.Marker(
    location=[42.3568701,-71.0682476],
    popup='Target',
    icon=folium.Icon(color='red', icon='ok-sign'),
).add_to(m)
folium.features.GeoJson(json.load(open("way_feature.json")),name="GeoJSON").add_to(m)
folium.LayerControl().add_to(m)
m.add_child(folium.LatLngPopup())
m
```
